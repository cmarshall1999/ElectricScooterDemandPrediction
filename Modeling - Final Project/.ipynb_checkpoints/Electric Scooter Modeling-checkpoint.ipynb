## LSTM Model

```python
num_x_signals = x_train.shape[1]
num_y_signals = y_train.shape[1]
num_train = x_train.shape[0]
```

```python
model = Sequential()
model.add(LSTM(25, input_shape = (1,num_x_signals)))
#model.add(LSTM(1, batch_input_shape = (109,1,num_x_signals),stateful=True))
model.add(Dense(1))
#model.add(Activation('tanh'))
#model.add(Dropout(0.3))
model.compile(loss='mse',optimizer='adam')
```

```python
x_train_reshape = x_train_scaled.reshape((x_train_scaled.shape[0], 1, x_train_scaled.shape[1]))
x_test_reshape = x_test_scaled.reshape((x_test_scaled.shape[0], 1, x_test_scaled.shape[1]))
```

```python
#fix NaN values before running this
history = model.fit(x_train_reshape,y_train_scaled,epochs=10000,batch_size=500)
```

```python
plt.plot(history.history['loss'], label='train')
#plt.plot(history.history['val_loss'], label='test')
plt.legend()
```
